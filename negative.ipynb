```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
conflict_df = pd.read_csv("african_conflicts.csv")
conflict_df
```

```python
intensity_occr = np.array(conflict_df.groupby(["Target Country"]).count()["Intensity"]).reshape(-1, 1)
scaler = MinMaxScaler()
normalized = scaler.fit_transform(intensity_occr)
data = np.concatenate((np.sort(conflict_df["Target Country"].unique()).reshape(-1, 1), normalized), axis=1)
data
```

```python
normal_df = pd.DataFrame(data, columns=["country", "normalized_intensity"])
normal_df
```

```python
normal_df.to_json("normal.json")
```
